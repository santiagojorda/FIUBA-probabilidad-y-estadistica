````python
import numpy as np
from scipy import special as sp
from scipy import stats
import math
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF

# DISCRETAS

def bernoulli(n, p, x):
  return sp.comb(n, x) * p**x * (1-p)**(n-x)

def binomial(n, p, x):
  return sp.comb(n, x) * p**x * (1-p)**(n-x)

def hipergeometrica(N, d, n, x):
  return sp.comb(d, x) * sp.comb(N-d, n-x) / sp.comb(N, n)

def poisson(u, x):
  return ((u**x) * (math.exp(-u)) ) / math.factorial(x)

# CONTINUAS

def exponencial(lamda, x):
  return gamma(1, lamda, x)


def _fun_gamma(n):
    return math.factorial(n-1)

def gamma(v, lamda, x):
  return ( (lamda**v) / _fun_gamma(v) * (x**(v-1)) * math.exp((-1) * lamda * x))

````

## **b)** Mediante 10 mil simulacioens estimar la probabilidad de que al arrojar 2 dados equilibrados la suma de los resultados sea menor que 11



````python
# Constantes y variables
NUM_SIMULACIONES = 10000
contador_sumas_menores_a_11 = 0

# Funciones
def lanzar_dado():
  return np.random.randint(1, 7)

def lanzar_y_sumar_dados():
  dado1 = lanzar_dado()
  dado2 = lanzar_dado()
  return dado1 + dado2

def es_menor_a_11(x):
  return x < 11

# Simulacion
for i in range(0, NUM_SIMULACIONES):
  suma = lanzar_y_sumar_dados()
  if(es_menor_a_11(suma)):
    contador_sumas_menores_a_11 += 1

# Imprimir contador
print(f"Cantidad de sumas menores a 11: {contador_sumas_menores_a_11}")

# Calcular la probabilidad
probabilidad_menor_a_11 = contador_sumas_menores_a_11 / NUM_SIMULACIONES
print(f"Probabilidad calculada de que la suma sea menor que 11: {probabilidad_menor_a_11}")
````

````python
# Constantes y variables
NUM_SIMULACIONES = 20
NUM_TIROS = 6
contador_experimento_con_as = 0

# Funciones
def lanzar_dado():
  return np.random.randint(0, 7)

def lanzar_6_dados():
  return [lanzar_dado() for _ in range(NUM_TIROS)]

# Simulacion
for i in range(0, NUM_SIMULACIONES):
    resultados = lanzar_6_dados()
    if 1 in resultados:
          contador_experimento_con_as += 1
    print(resultados)


# Imprimir cantidad de ases
print(f"Cantidad de ases: {contador_experimento_con_as}")

# Calcular probabilidad
probabilidad_de_ases = contador_experimento_con_as / NUM_SIMULACIONES
print(f"Probabilidad calculada de que salgan ases: {probabilidad_de_ases}")
````

## a) Estimar la probabilidad de que el precio de un kilo de asado supere los 9.5

````python
# Se obtinen 24 precios al azar del kilo de asado en distintas carnicerias
resultados = [8.59, 8.02, 8.77, 10.35, 8.29, 7.15, 7.50, 9.00, 9.18, 9.15, 8.53, 9.11, 10.03, 7.63, 8.97, 9.66, 8.47, 10.20, 9.98, 9.01, 9.00, 8.73, 9.44, 10.54]

# se ordenan para analizarlo mejor visualmente
resultados.sort()
print(resultados)

# obtengo el total
length = len(resultados)
print(f"Cantidad de resultados: {len(resultados)}")

PRECIO_ESTIMADO = 9.5

def estimar_probabilidad_precio_superior_a(precio_estimado):

  cant_superiores = 0

  # cuenta la cantidad de precios superiores a precio_estimado
  for resultado in resultados:
    if(resultado > precio_estimado):
      cant_superiores += 1

  probabilidad = cant_superiores / length
  print(f"Cantidad de precios superiores a {PRECIO_ESTIMADO} es {cant_superiores}")
  return probabilidad

print(f"La probabilidad de que el precio sea superior a {PRECIO_ESTIMADO}$: ", estimar_probabilidad_precio_superior_a(PRECIO_ESTIMADO))
````

# **Ejercicio 6.3**

### Se arroja una moneda equilibrada 18 veces

### Hallar el numero mas probable de caras, y calcular la probabilidad de obtenerlo


````python
cant_tiros = 18
prob_cara = 0.5

maximo = [0, 0]

for n in range(cant_tiros):
  prob_obtenida = bernoulli(cant_tiros, prob_cara, n+1)
  if prob_obtenida > maximo[1]:
    maximo = [n+1, prob_obtenida]
  print(f"para {n+1} caras, la probabilidad es: {prob_obtenida}")

print('\n')
print(f"el numero de caras que tiene un max de probabilidad es el {maximo[0]}, con un {round(maximo[1]*100, 2)}%")
````
