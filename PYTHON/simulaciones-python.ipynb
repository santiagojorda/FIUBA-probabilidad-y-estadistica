# EXPONENCIAL

```python
distribucion = stats.expon(scale = 1) # func de distribucion lamda = 1
numero_de_tiradas = 2000
num_aleatorios = np.random.rand(numero_de_tiradas)

exponencial = - np.log(1.0 - num_aleatorios) # log = logaritmo natural = np
print(exponencial)

extremos = [0, 0.2, 0.4, 0.6, 0.8, 1, 2, 4]
x = np.linspace(distribucion.ppf(0.01), distribucion.ppf(0.99), 100)
plt.hist(exponencial, bins=extremos, density=True, label='Histograma')
plt.plot(x, distribucion.pdf(x), label='Densidad')
plt.legend()
plt.show()
```

# **BINOMIAL A TRAVES DE UNA NORMAL**

tiro una moneda equilibrada 3 veces
X es una binomial(3, p=0.5)

busco P(X<=x) , x = {0, 1, 2, 3}

```python
binomial = stats.binom(n=3, p=0.5)
print('Fx(x) = P(X<=x) = cdf')
print('P(X<=0): ', binomial.cdf(0)) #cdf comulative distribution function Fx(x) = P(X<=x)
print('P(X<=1): ', binomial.cdf(1))
print('P(X<=2): ', binomial.cdf(2))
print('P(X<=3): ', binomial.cdf(3))

print()

normal = stats.norm()
print('P(X<=a) = %    => cuantiles: a = ppf') # ppf Percent Point Function = cuantil
print('P(X<=a) = 0.9  => a =', normal.ppf(0.9))
print('P(X<=a) = 0.99  => a =', normal.ppf(0.99))

muestra_normal = normal.rvs(size=100) #rvs muestra aleatoria
```

**POISSON**

```python
mu = 10
dist = stats.poisson(mu)

# cdf P(X<=x)
# pmf P(X=x)
# sf P(X>x)
# E[x]
# DISCRETA
# E[x| x <= 4]
# print(sum(x * dist.pmf(x) for x in range(5))) # E[x| x <= 4]

print(dist.cdf(1)) # cdf P(X<=x)
print(dist.pmf(1)) # pmf P(X=x)
print(dist.sf(1)) # sf P(X>x)
print(dist.mean()) # E[x]
print()

probabilidad = dist.cdf(4)
print('P(X<=4)', probabilidad)

esperanza_condicional = sum(x * dist.pmf(x) for x in range(5))
print('E[X | X<=4]', esperanza_condicional) # E[x| x <= 4]

# E[X | X<=4]
print('E[X | X<=4] / P(X<=4)' , esperanza_condicional / probabilidad )
```
